An "index" refers to the position of a data element (or elements) inside a collection of data.

Examples:
* Indexing a list: `a[0]` uses the index 0 to get the first element out of the list a
* Indexing a dictionary: `a['first']` uses the index 'first' to get the element with key 'first' out of the dictionary a
* Indexing a dataframe:
  * `a['State']` uses the index 'State' (a column name) to get all the elements in the column 'State ' out of the dataframe a
  * `a.loc[0:10, 'State']` uses the indices 0:10 to get the first 10 rows and the index 'State' to get only the 'State' column
  * `a.loc[0:10, ['State','Year']]` uses the indices 0:10 to get the first 10 rows and the list ['State','Year'] to get the two columns indexed by 'State' and 'Year'
