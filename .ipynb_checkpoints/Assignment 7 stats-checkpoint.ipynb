```python
#Tokenization
```

```python
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
sent = sent_tokenize(response.text)
print(sent)
```

```python
sent[0]
```

```python
print(word_tokenize(sent[1]))
```

```python
words = []
for s in sent:
    for w in word_tokenize(s):
        words.append(w)
```

```python
print(words)
#This is howI Tokenize the words in the text Othello
```

```python
#Stop Word removal
```

```python
from nltk.corpus import stopwords
from string import punctuation
nltk.download('stopwords')
```

```python
print(punctuation)
```

```python
print(stopwords.words('english'))
```

```python
print(list(punctuation))
```

```python
myStopWords = list(punctuation) + stopwords.words('english')
```

```python
print(words)
```

```python
wordsNoStop = []
for i in words:
    if i not in myStopWords:
        wordsNoStop.append(i)
print(words)
print(wordsNoStop)
```

```python
wordsNoStopComp = [w for w in words if w not in myStopWords]
print(wordsNoStopComp)
```

```python
#Setting up linear regression
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
X = features_df[['bmi']]
y = features_df['target']
#Here I am setting linear regression between bmi and the target
#bmi is the variable with the highest corelation with target
```

```python
reg = LinearRegression().fit(X, y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
ytrain = reg.intercept_ + reg.coef_ * X
```

```python
plt.plot(X,y,'ro',X,ytrain,'b-');
```

```python
#This graph demonstrates a positive correlation between bmi and the target
#The higher the bmi the faster the progression of disease
```

```python
mean_squared_error(y, ytrain)
```

```python
r2_score(y, ytrain)
```

```python
X = features_df[['sex']]
y = features_df['target']
```

```python
reg = LinearRegression().fit(X, y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
ytrain = reg.intercept_ + reg.coef_ * X
```

```python
plt.plot(X,y,'ro',X,ytrain,'b-');
```

```python
#Sex has the lowest correlation with the target.
#The graph reflects this as well, there isn't a correlation between sex and progresssion of disease.
```

```python
mean_squared_error(y, ytrain)
```

```python
r2_score(y, ytrain)
```
