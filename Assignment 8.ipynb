```python
characters = [
'DUKE OF VENICE',
'BRABANTIO',
'GRATIANO'
'LODOVICO', 
'OTHELLO', 
'CASSIO',
'IAGO', 
'MONTANO', 
'RODERIGO', 
'CLOWN', 
'DESDEMONA', 
'EMILIA', 
'BIANCA', 
'Officers',
'Gentlemen',
'Messenger',
'Musicians',
'Herald',
'Sailor',
'Attendants',
]
```

```python
len(acts)
```

```python
charnum = {}

connections = {}
for i in range(len(characters)-1):
    for j in range(i+1,len(characters)):
        connections[(characters[i],characters[j])] = 0

for k in characters:
    charnum[k] = 0

for i in acts:
    for j in i.split('SCENE')[1:]:
        scenechars = []
        for k in characters:
            if j.find(k) != -1:
                scenechars.append(k)
                charnum[k] += 1
        for a in range(len(scenechars)-1):
            for b in range(a+1,len(scenechars)):
                connections[(scenechars[a],scenechars[b])] += 1
```

```python
from networkx.algorithms.community.centrality import girvan_newman
```

```python
communities = girvan_newman(G)
```

```python
node_groups = []
for com in next(communities):
    node_groups.append(list(com))
 
print(node_groups)
 
color_map = []
for node in G:
    if node in node_groups[0]:
        color_map.append('blue')
    else:
        color_map.append('green')
nx.draw(G, node_color=color_map, with_labels=True)
plt.show()
```

```python
#All the nodes belong to the same group except for the musicians.
#This graph is made with the shortest distance between Othello and Casio in mind.
#The closer together nodes are the
#The musicians share a scene with the clown (Edge connection) The musicians have a second degree connection with
#Bianca. This means that the musicians aren't in a scene with Bianca they just have a connection with someone
#That's in a scene with Bianca.
```

```python
from networkx.algorithms.community import greedy_modularity_communities
```

```python
greedy_modularity_communities(G)
```

```python
communities = greedy_modularity_communities(G)
# Create empty dictionary
modularity_class = {}
#Loop through each community in the network
for community_number, community in enumerate(communities):
    #For each member of the community, add their community number
    for name in community:
        modularity_class[name] = community_number
```

```python
modularity_class
```

```python
communities = greedy_modularity_communities(G)

node_groups = []
for com in communities:
    node_groups.append(list(com))
 
print(node_groups)
 
color_map = []
for node in G:
    if node in node_groups[0]:
        color_map.append('blue')
    elif node in node_groups[1]:
        color_map.append('yellow')
    else:
        color_map.append('green')
nx.draw(G, node_color=color_map, with_labels=True)
plt.show()
```

```python
#There are 2 groups present in this graph.
#The blue group is more closely associated with Othello
#The yellow group is more closely associated with Cassio and Iago.
#The algorithim can't tell who is friends or enemies with who.
#However, the algorithim could tell which characters shares scence with one another.
```

```python
graphitems = []

for k,v in charnum.items():
    if k in node_groups[0]:
        modularity = 'blue'
    elif k in node_groups[1]:
        modularity = 'green'
    else:
        modularity = 'red'
    dashnode = {'data': {'id': k,
                         'label': k.title(),
                         'size': str(v),'modularity': modularity}}
    graphitems.append(dashnode)

correction_factor = max(connections.values())
print(correction_factor)
for k,v in connections.items():
    if v != 0:
        dashedge = {'data': {'source': k[0],
                             'target': k[1],
                             'weight': v/correction_factor}}
        graphitems.append(dashedge)
    
graphitems
```
